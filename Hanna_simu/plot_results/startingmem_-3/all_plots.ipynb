```python

data = epi_1
start = 150

# Extract variables from the loaded dictionary
meanw = data['meanw'][:,:-1]
meanmemory_p = data['meanmemory_p'][:,:-1]
meanneutral_p = data['meanneutral_p'][:,:-1]

rho= data['rho_m_alpha_beta'][:,0].flatten()
m= data['rho_m_alpha_beta'][:,1].flatten()

alpha = data['rho_m_alpha_beta'][:,2].flatten()
beta = data['rho_m_alpha_beta'][:,3].flatten()
nof_scenarios = meanw.shape[0]
maxgen = meanw.shape[1]

# Loop over scenarios
for i in range(start, nof_scenarios):  
    
    plt.figure(3)
    plt.clf()  # clear previous figure
    
    # Create two vertically stacked subplots
    fig, axs = plt.subplots(2, 1, figsize=(8, 6))
    plt.subplots_adjust(hspace=0.4)
    
    # First plot: mean fitness
    axs[0].plot(meanw[i, :])
    axs[0].set_ylabel('Mean fitness')
    axs[0].axis([1, maxgen, 0, 1])
     
    # Second plot: mean memory
    axs[1].plot(meanmemory_p[i, :])
    axs[1].plot(meanneutral_p[i, :])
    axs[1].set_ylabel('Mean memory')
    axs[1].set_xlabel('Generation')
    axs[1].axis([1, maxgen, 0, 1])
    axs[1].legend()
    # Title with parameters
    fig.suptitle(f" = {rho[i]:.3f}, m = {m[i]:.3f},  = {alpha[i]:.3f},  = {beta[i]:.3f}")
    
    plt.show()
    input("Press Enter to continue...")  # equivalent to MATLAB's 'pause'
```
