## I borrow aspects of the genetic-epigenetic structure, fitness functions and environmental heterogeneity from the 2 papers

### Genetic-epigenetic cue and maternal phenotype structure

The adult phenotype of an individual is given by  
$$
x = \alpha z + \beta_J c_J + \beta_A c_A + \gamma (m + \epsilon_m) + \delta \epsilon_{\delta}
$$

where:

$$
m = x_{t-1}
$$

Genetic 'cue' genes:

$$
z = z' + \epsilon_Z \quad \text{where} \quad \epsilon_Z \sim \mathcal{N}(0, \sigma_{\text{mut}}^2)
$$

Juvenile cue:

$$
c_J \sim \mathcal{N}(\theta(t), \sigma_J^2)
$$

Adult cue:

$$
c_A \sim \mathcal{N}(\theta(t-1), \sigma_A^2)
$$

Maternal noise:

$$
\epsilon_m \sim \mathcal{N}(0, \sigma_m^2)
$$

Developmental noise:

$$
\epsilon_{\delta} \sim \mathcal{N}(0, 1)
$$


### environment

$$
\theta(t + 1) = \lambda.\theta(t) + \epsilon_{\theta}
$$


 The number of surviving offspring for an individual with phenotype (x) in environment (h) is:
 
$$
K \exp\left(-\frac{1}{2}(x - \theta)^2\right)
$$




### Genetic-epigenetic cue and maternal phenotype structure from the simpler epigenetic resetting paper









$$
m_t = m_0 + m_1(x_t + \epsilon_t) ; \quad \text{with} \quad \epsilon_t \sim \mathcal{N}(0, \epsilon_{\sigma})

So m0 is the baseline maternal effect if the mother does not respond to her environment, and m1 describes how much the maternal effect changes for a change in one unit of the environmental state.

The process is defined as:

\begin{equation}
y_t = y_{t-1} + \epsilon_Y \quad \text{where} \quad \epsilon_Y \sim \mathcal{N}(0, \sigma_{\text{epimut}}^2)
\end{equation}

This represents a \textbf{random walk}, where the innovation term \( \epsilon_Y \) is normally distributed with mean 0 and variance \( \sigma_{\text{epimut}}^2 \).

Starting from an initial value \( y_0 \), the process evolves as:

\begin{equation}
y_t = y_0 + \sum_{i=1}^{t} \epsilon_{Y,i}
\end{equation}

Hence, the distribution of \( y_t \) is:

\begin{equation}
y_t \sim \mathcal{N}\left(y_0, t \cdot \sigma_{\text{epimut}}^2\right)
\end{equation}
